```python
import tifffile as tiff

def load_img(idx):
    img = tiff.imread(os.path.join("train_images", f"{idx}.tiff"))
    return img
```

```python
import tensorflow as tf
import numpy as np

def rle_decode(rle_mask, shape):

    rle_mask = rle_mask.split()
    starts, lengths = [np.asarray(x, dtype=int) for x in (rle_mask[0:][::2], rle_mask[1:][::2])]
    starts -= 1
    ends = starts + lengths

    mask = np.zeros((shape[0] * shape[1]), dtype=np.uint8)
    for start, end in zip(starts, ends):
        mask[start:end] = 1

    mask = mask.reshape(shape[0], shape[1]).T
    return mask[:,:,tf.newaxis]
```

```python
import pandas as pd

df = pd.read_csv("my_train.csv",index_col="id")
df['organ'] = pd.Categorical(df['organ'], ['prostate', 'spleen', 'lung', 'kidney', 'largeintestine'])
```

```python
import os
train_img_output_path = './images_png'
train_mask_output_path = './masks_png'
if not os.path.exists(train_img_output_path):
    os.makedirs(train_img_output_path)
if not os.path.exists(train_mask_output_path):
    os.makedirs(train_mask_output_path)
```

```python
from keras.utils import save_img

for idx, row in df.iterrows():
    img = load_img(idx)
    save_img(f'{train_img_output_path}/{idx}.png', img)
    mask = rle_decode(row["rle"], (row["img_height"],row["img_width"]))
    save_img(f'{train_mask_output_path}/{idx}.png', mask, scale=False)
```
